# sanitize

```python
# doesn't have 'name' wtf
list(filter(lambda x: x['tags'].get('name') == None, rawRes["elements"]))
```

```python
rawRes_Sane = rawRes
rawRes_Sane['elements'] = list(filter(lambda x: x['id'] != 17893356, rawRes["elements"]))
len(rawRes_Sane['elements'])
```

```python
{ele['id']:ele['tags']['name'] for ele in rawRes_Sane['elements']}
```

# normalize

```python
countrIdDict = {ele['id']:next((ele['tags'][k] for k in ['name:en','name'] if k in ele['tags'])) for ele in rawRes_Sane['elements']}
len(countrIdDict)
```

```python
dump('./data/countrIdDict.json', countrIdDict)
```
